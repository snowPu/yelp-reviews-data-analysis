```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

with open('/Users/jieyizhang/Downloads/yelp_dataset/yelp_academic_dataset_review.json') as json_file:      
    data = json_file.readlines()
    data = list(map(json.loads, data)) 

review_df = pd.DataFrame(data)


with open('/Users/jieyizhang/Downloads/yelp_dataset/yelp_academic_dataset_user.json') as json_file:      
    data = json_file.readlines()
    data = list(map(json.loads, data)) 

user_df = pd.DataFrame(data)
```

```python
reviewnum_user = review_df.groupby('user_id')['review_id'].count().to_frame()
reviewnum_user = reviewnum_user.reset_index()
```

```python
reviewnum_user.columns = ['user_id', 'review_count(R)']
```

```python
large_df = user_df.merge(reviewnum_user,left_on='user_id',right_on='user_id',how='outer')
```

```python
#large_df
large_df.to_json("/Users/jieyizhang/Downloads/yelp_dataset/yelp_academic_dataset_UserReviewNum.json",orient='split')
```

```python
print('The number of instances in User outer joined table Reviewnum_user on user_id is', str(len(large_df)))
print('The numbber of instances in User Table is',str(len(user_df)))
print('The numbber of instances in ReviewNum User Table is',str(len(reviewnum_user)))
```

```python
Correct_Users = large_df[large_df['review_count']==large_df['review_count(R)']]

print('The number of users whose review count are consistent in User_table and Review table is : ',str(len(Correct_Users)))
print('The percentage of users whose review count are consistent in User_table and Review table is : ',str(len(Correct_Users)/len(large_df)*100),'%')

```
